```python
# ============================================================
# KLASİK SVM KARŞILAŞTIRMA MODELİ
# İlk yüklenen daha iyi sonuç veren modeldeki klasik SVM akışına göre düzenlendi
# Hibrit QK-Gated SVM'deki klasik baseline ile uyumlu yapı:
# EfficientNet-B0 + PCA(256) + RBF-SVM
# Sınıf bazlı doğru / yanlış tahmin tablosu dahil
# ============================================================

from google.colab import drive
drive.mount("/content/drive")

# ============================================================
# 1) KURULUMLAR
# ============================================================
!pip -q install --upgrade pip
!pip -q install torchvision scikit-learn pandas tqdm matplotlib joblib openpyxl

# ============================================================
# 2) IMPORTLAR
# ============================================================
import os
import random
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
from PIL import Image
from tqdm.auto import tqdm

import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader

import torchvision.transforms as T
import torchvision.models as models

import matplotlib.pyplot as plt
import joblib

from sklearn.model_selection import train_test_split, GridSearchCV, StratifiedKFold
from sklearn.preprocessing import StandardScaler, label_binarize
from sklearn.decomposition import PCA
from sklearn.svm import SVC
from sklearn.pipeline import Pipeline
from sklearn.metrics import (
    accuracy_score,
    f1_score,
    roc_auc_score,
    classification_report,
    confusion_matrix,
    ConfusionMatrixDisplay,
    roc_curve,
    auc
)

# ============================================================
# 3) GENEL AYARLAR
# ============================================================
SEED = 42

def seed_everything(seed=42):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)

    torch.backends.cudnn.deterministic = False
    torch.backends.cudnn.benchmark = True

seed_everything(SEED)

DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Kullanılan cihaz:", DEVICE)

DATA_ROOT = "/content/drive/MyDrive/Beyin_MR_Kuantum_Makale/Data"

CLASS_NAMES = [
    "glioma_tumor",
    "meningioma_tumor",
    "no_tumor",
    "pituitary_tumor"
]

CLASS_TO_IDX = {cls: i for i, cls in enumerate(CLASS_NAMES)}
IDX_TO_CLASS = {i: cls for cls, i in CLASS_TO_IDX.items()}

N_CLASSES = len(CLASS_NAMES)

print("Sınıf eşlemesi:", CLASS_TO_IDX)

IMG_SIZE = 224
TEST_SIZE = 0.20

FEATURE_MODEL_NAME = "efficientnet_b0"
FEATURE_BATCH_SIZE = 64

PCA_COMPONENTS = 256

# İlk yüklenen daha iyi sonuç veren modeldeki klasik SVM baseline ile aynı arama uzayı
SVM_PARAM_GRID_CLASSICAL = {
    "svm__C": [1, 3, 10, 30, 100],
    "svm__gamma": ["scale", 0.001, 0.003, 0.006, 0.01, 0.03]
}

CV_SPLITS = 5
SVM_KERNEL = "rbf"

FORCE_REEXTRACT_IMAGE_FEATURES = False

# Hibrit modelin cache klasörü
HYBRID_CACHE_DIR = "/content/drive/MyDrive/Beyin_MR_Kuantum_Makale/cache_stronger_quantum_kernel_svm"
os.makedirs(HYBRID_CACHE_DIR, exist_ok=True)

# Klasik SVM modelin kendi cache klasörü
CACHE_DIR = "/content/drive/MyDrive/Beyin_MR_Kuantum_Makale/cache_classical_svm_gridsearch_stronger"
os.makedirs(CACHE_DIR, exist_ok=True)

HYBRID_TRAIN_SPLIT_PATH = os.path.join(HYBRID_CACHE_DIR, "train_split.csv")
HYBRID_TEST_SPLIT_PATH = os.path.join(HYBRID_CACHE_DIR, "test_split.csv")

SVM_TRAIN_SPLIT_PATH = os.path.join(CACHE_DIR, "train_split.csv")
SVM_TEST_SPLIT_PATH = os.path.join(CACHE_DIR, "test_split.csv")

HYBRID_RAW_FEATURE_CACHE_PATH = os.path.join(
    HYBRID_CACHE_DIR,
    f"image_features_{FEATURE_MODEL_NAME}_80_20_seed{SEED}_suball_img{IMG_SIZE}.npz"
)

SVM_FEATURE_CACHE_PATH = os.path.join(
    CACHE_DIR,
    f"svm_efficientnet_b0_features_seed{SEED}_img{IMG_SIZE}.npz"
)

MODEL_SAVE_PATH = os.path.join(CACHE_DIR, "classical_svm_gridsearch_stronger_model.joblib")
DETAIL_TEST_RESULTS_PATH = os.path.join(CACHE_DIR, "svm_goruntu_bazli_test_sonuclari.csv")
CLASS_SUMMARY_PATH = os.path.join(CACHE_DIR, "svm_sinif_bazli_dogru_yanlis.csv")
CONFUSION_DETAIL_PATH = os.path.join(CACHE_DIR, "svm_siniflar_arasi_karisiklik.csv")

# ============================================================
# 4) VERİ LİSTELEME VE HİBRİT MODELLE AYNI 80/20 BÖLME
# ============================================================
def list_images_recursive(folder):
    exts = (".png", ".jpg", ".jpeg", ".bmp", ".webp", ".tif", ".tiff")
    files = []

    if not os.path.isdir(folder):
        return files

    for root, _, names in os.walk(folder):
        for name in names:
            if name.lower().endswith(exts):
                files.append(os.path.join(root, name))

    return sorted(files)


def collect_all_class_files(data_root, class_names):
    """
    Şu veri düzenlerini destekler:

    1) Data/glioma_tumor
       Data/meningioma_tumor
       Data/no_tumor
       Data/pituitary_tumor

    2) Data/Training/glioma_tumor
       Data/Testing/glioma_tumor
       ...

    Amaç: Tüm görüntüleri toplayıp aynı 80/20 bölme mantığıyla ayırmaktır.
    """

    class_files = {}

    for cls in class_names:
        candidates = [
            os.path.join(data_root, cls),
            os.path.join(data_root, "Training", cls),
            os.path.join(data_root, "Testing", cls)
        ]

        files = []

        for folder in candidates:
            files.extend(list_images_recursive(folder))

        files = sorted(list(set(files)))

        if len(files) == 0:
            raise ValueError(f"Bu sınıf için görüntü bulunamadı: {cls}")

        class_files[cls] = files

    return class_files


def build_classwise_split(data_root, class_names, test_size=0.2, seed=42):
    class_files = collect_all_class_files(data_root, class_names)

    train_items = []
    test_items = []
    all_items = []

    for cls in class_names:
        files = class_files[cls]
        y = CLASS_TO_IDX[cls]

        tr, te = train_test_split(
            files,
            test_size=test_size,
            random_state=seed,
            shuffle=True
        )

        train_items.extend([(str(p), y, cls) for p in tr])
        test_items.extend([(str(p), y, cls) for p in te])
        all_items.extend([(str(p), y, cls) for p in files])

    rng = np.random.default_rng(seed)
    rng.shuffle(train_items)
    rng.shuffle(test_items)
    rng.shuffle(all_items)

    return train_items, test_items, all_items


def save_split_csv(items, path):
    df = pd.DataFrame(items, columns=["path", "label", "class_name"])
    df.to_csv(path, index=False)


def load_split_csv(path):
    df = pd.read_csv(path)

    items = []

    for _, row in df.iterrows():
        image_path = str(row["path"])

        if "class_name" in df.columns:
            class_name = str(row["class_name"])
            label = CLASS_TO_IDX[class_name]
        else:
            label = int(row["label"])
            class_name = IDX_TO_CLASS[label]

        items.append((image_path, label, class_name))

    return items


def load_hybrid_split_if_exists():
    if os.path.exists(HYBRID_TRAIN_SPLIT_PATH) and os.path.exists(HYBRID_TEST_SPLIT_PATH):
        print("\nHibrit modelde kullanılan train/test split dosyaları bulundu.")
        train_items = load_split_csv(HYBRID_TRAIN_SPLIT_PATH)
        test_items = load_split_csv(HYBRID_TEST_SPLIT_PATH)
        all_items = train_items + test_items
        return train_items, test_items, all_items

    return None, None, None


train_items, test_items, all_items = load_hybrid_split_if_exists()

if train_items is None:
    if os.path.exists(SVM_TRAIN_SPLIT_PATH) and os.path.exists(SVM_TEST_SPLIT_PATH):
        print("\nSVM için kayıtlı train/test split dosyaları bulundu.")
        train_items = load_split_csv(SVM_TRAIN_SPLIT_PATH)
        test_items = load_split_csv(SVM_TEST_SPLIT_PATH)
        all_items = train_items + test_items
    else:
        print("\nHibrit split bulunamadı. Tüm veri üzerinden yeni 80/20 split oluşturuluyor.")
        train_items, test_items, all_items = build_classwise_split(
            DATA_ROOT,
            CLASS_NAMES,
            test_size=TEST_SIZE,
            seed=SEED
        )

        save_split_csv(train_items, SVM_TRAIN_SPLIT_PATH)
        save_split_csv(test_items, SVM_TEST_SPLIT_PATH)

print("\nVeri özeti:")
print("Toplam:", len(all_items))
print("Eğitim:", len(train_items))
print("Test  :", len(test_items))


def print_distribution(name, items):
    y = np.array([i[1] for i in items])

    print(f"\n{name} dağılımı:")
    for i, cls in enumerate(CLASS_NAMES):
        count = int(np.sum(y == i))
        print(f"{cls:18s}: {count}")


print_distribution("Eğitim", train_items)
print_distribution("Test", test_items)

# ============================================================
# 5) EFFICIENTNET-B0 İLE 1280 BOYUTLU ÖZELLİK ÇIKARIMI
# ============================================================
feature_transform = T.Compose([
    T.Resize((IMG_SIZE, IMG_SIZE)),
    T.Grayscale(num_output_channels=3),
    T.ToTensor(),
    T.Normalize(
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    )
])


class ImagePathDataset(Dataset):
    def __init__(self, items, transform):
        self.items = items
        self.transform = transform

    def __len__(self):
        return len(self.items)

    def __getitem__(self, idx):
        path, y, _ = self.items[idx]
        img = Image.open(path).convert("RGB")
        x = self.transform(img)
        return x, y


def build_efficientnet_b0_feature_extractor():
    weights = models.EfficientNet_B0_Weights.DEFAULT
    model = models.efficientnet_b0(weights=weights)

    # Classifier kaldırılır.
    # EfficientNet-B0 artık 1280 boyutlu özellik vektörü üretir.
    model.classifier = nn.Identity()

    model = model.to(DEVICE)
    model.eval()

    for p in model.parameters():
        p.requires_grad = False

    return model, 1280


feature_extractor, FEATURE_DIM = build_efficientnet_b0_feature_extractor()

print("\nFeature extractor:", FEATURE_MODEL_NAME)
print("Ham öznitelik boyutu:", FEATURE_DIM)


@torch.no_grad()
def extract_image_features(items, split_name):
    dataset = ImagePathDataset(items, feature_transform)

    loader = DataLoader(
        dataset,
        batch_size=FEATURE_BATCH_SIZE,
        shuffle=False,
        num_workers=2,
        pin_memory=(DEVICE.type == "cuda")
    )

    features = []
    labels = []

    for x, y in tqdm(loader, desc=f"{split_name} EfficientNet-B0 özellikleri çıkarılıyor"):
        x = x.to(DEVICE, non_blocking=True)

        f = feature_extractor(x)
        f = f.detach().cpu().numpy().astype(np.float32)

        features.append(f)
        labels.append(np.asarray(y, dtype=np.int64))

    X = np.concatenate(features, axis=0)
    y = np.concatenate(labels, axis=0)

    return X, y


using_hybrid_features = (
    os.path.exists(HYBRID_RAW_FEATURE_CACHE_PATH)
    and os.path.exists(HYBRID_TRAIN_SPLIT_PATH)
    and os.path.exists(HYBRID_TEST_SPLIT_PATH)
    and not FORCE_REEXTRACT_IMAGE_FEATURES
)

if using_hybrid_features:
    print("\nHibrit modelde kullanılan kayıtlı EfficientNet-B0 özellikleri yükleniyor:")
    print(HYBRID_RAW_FEATURE_CACHE_PATH)

    data = np.load(HYBRID_RAW_FEATURE_CACHE_PATH)

    X_train_raw = data["X_train_raw"]
    y_train = data["y_train"]
    X_test_raw = data["X_test_raw"]
    y_test = data["y_test"]

elif os.path.exists(SVM_FEATURE_CACHE_PATH) and not FORCE_REEXTRACT_IMAGE_FEATURES:
    print("\nSVM için kayıtlı EfficientNet-B0 özellikleri yükleniyor:")
    print(SVM_FEATURE_CACHE_PATH)

    data = np.load(SVM_FEATURE_CACHE_PATH)

    X_train_raw = data["X_train_raw"]
    y_train = data["y_train"]
    X_test_raw = data["X_test_raw"]
    y_test = data["y_test"]

else:
    X_train_raw, y_train = extract_image_features(train_items, "Eğitim")
    X_test_raw, y_test = extract_image_features(test_items, "Test")

    np.savez_compressed(
        SVM_FEATURE_CACHE_PATH,
        X_train_raw=X_train_raw,
        y_train=y_train,
        X_test_raw=X_test_raw,
        y_test=y_test
    )

    print("\nEfficientNet-B0 özellikleri kaydedildi:")
    print(SVM_FEATURE_CACHE_PATH)

print("\nHam özellik şekilleri:")
print("X_train_raw:", X_train_raw.shape)
print("X_test_raw :", X_test_raw.shape)

# ============================================================
# 6) STANDARD SCALER + PCA(256)
# ============================================================
raw_scaler = StandardScaler()

X_train_scaled = raw_scaler.fit_transform(X_train_raw).astype(np.float32)
X_test_scaled = raw_scaler.transform(X_test_raw).astype(np.float32)

pca_dim = min(
    PCA_COMPONENTS,
    X_train_scaled.shape[0] - 1,
    X_train_scaled.shape[1]
)

pca = PCA(
    n_components=pca_dim,
    random_state=SEED,
    svd_solver="randomized"
)

X_train_pca = pca.fit_transform(X_train_scaled).astype(np.float32)
X_test_pca = pca.transform(X_test_scaled).astype(np.float32)

explained_var = float(np.sum(pca.explained_variance_ratio_))

print("\nPCA sonrası özellik şekilleri:")
print("X_train_pca:", X_train_pca.shape)
print("X_test_pca :", X_test_pca.shape)
print(f"PCA açıklanan varyans: {explained_var:.4f}")

# ============================================================
# 7) KLASİK RBF-SVM MODELİ
# İlk yüklenen güçlü modeldeki klasik SVM baseline ile aynı yapı:
# PCA(256) sonrası tekrar StandardScaler + GridSearchCV + RBF-SVM
# ============================================================
print("\n===================================================")
print("GRIDSEARCHCV TABANLI KLASİK RBF-SVM EĞİTİMİ BAŞLIYOR")
print("===================================================")

X_train_classical = X_train_pca
X_test_classical = X_test_pca

cv = StratifiedKFold(
    n_splits=CV_SPLITS,
    shuffle=True,
    random_state=SEED
)

classical_pipeline = Pipeline([
    ("scaler", StandardScaler()),
    ("svm", SVC(
        kernel=SVM_KERNEL,
        probability=True,
        class_weight="balanced",
        random_state=SEED
    ))
])

classical_grid = GridSearchCV(
    classical_pipeline,
    SVM_PARAM_GRID_CLASSICAL,
    scoring="f1_macro",
    cv=cv,
    n_jobs=-1,
    verbose=2,
    refit=True
)

print("Aranan SVM parametreleri:")
print(SVM_PARAM_GRID_CLASSICAL)
print(f"Çapraz doğrulama: {CV_SPLITS} katlı StratifiedKFold")
print("Başarı ölçütü: Macro-F1")

classical_grid.fit(X_train_classical, y_train)

svm_model = classical_grid.best_estimator_
best_params = classical_grid.best_params_
best_cv_macro_f1 = float(classical_grid.best_score_)

print("\nSVM eğitimi tamamlandı.")
print("En iyi SVM parametreleri:", best_params)
print(f"En iyi CV Macro-F1: {best_cv_macro_f1:.4f}")

joblib.dump(
    {
        "model": svm_model,
        "grid_search": classical_grid,
        "raw_scaler": raw_scaler,
        "pca": pca,
        "class_names": CLASS_NAMES,
        "class_to_idx": CLASS_TO_IDX,
        "idx_to_class": IDX_TO_CLASS,
        "pca_dim": X_train_pca.shape[1],
        "pca_explained_variance": explained_var,
        "svm_param_grid": SVM_PARAM_GRID_CLASSICAL,
        "best_params": best_params,
        "best_cv_macro_f1": best_cv_macro_f1,
        "svm_kernel": SVM_KERNEL,
        "cv_splits": CV_SPLITS,
        "scoring": "f1_macro"
    },
    MODEL_SAVE_PATH
)

print("\nModel kaydedildi:")
print(MODEL_SAVE_PATH)

# 8) TEST DEĞERLENDİRMESİ
# ============================================================
def safe_macro_auc(y_true, y_proba):
    try:
        return roc_auc_score(
            y_true,
            y_proba,
            multi_class="ovr",
            average="macro"
        )
    except Exception:
        return np.nan


# İlk yüklenen güçlü modeldeki klasik SVM baseline ile aynı şekilde
# tahminler predict_proba çıktısının argmax değeriyle üretilir.
y_proba = svm_model.predict_proba(X_test_classical)
y_pred = np.argmax(y_proba, axis=1)

accuracy = accuracy_score(y_test, y_pred)
macro_f1 = f1_score(y_test, y_pred, average="macro", zero_division=0)
weighted_f1 = f1_score(y_test, y_pred, average="weighted", zero_division=0)
macro_auc = safe_macro_auc(y_test, y_proba)

test_total = len(y_test)
correct_total = int(np.sum(y_pred == y_test))
wrong_total = int(np.sum(y_pred != y_test))

print("\n===================================================")
print("KLASİK SVM TEST SONUÇLARI")
print("===================================================")
print(f"Accuracy    : {accuracy:.4f}")
print(f"Macro-F1    : {macro_f1:.4f}")
print(f"Weighted-F1 : {weighted_f1:.4f}")
print(f"Macro-AUC   : {macro_auc:.4f}")
print(f"Test toplam : {test_total}")
print(f"Doğru tahmin: {correct_total}")
print(f"Yanlış tahmin: {wrong_total}")
print(f"Best params : {best_params}")
print(f"CV Macro-F1 : {best_cv_macro_f1:.4f}")

general_summary_df = pd.DataFrame([{
    "Model": "Klasik SVM: EfficientNet-B0 + PCA(256) + GridSearchCV RBF-SVM",
    "Best Params": str(best_params),
    "CV Macro-F1": best_cv_macro_f1,
    "Test toplam": test_total,
    "Doğru tahmin": correct_total,
    "Yanlış tahmin": wrong_total,
    "Accuracy": accuracy,
    "Macro-F1": macro_f1,
    "Weighted-F1": weighted_f1,
    "Macro-AUC": macro_auc
}])

print("\n===================================================")
print("GENEL TEST SONUÇ TABLOSU")
print("===================================================")
display(general_summary_df.round(4))

# ============================================================
# 9) SINIF BAZLI DOĞRU / YANLIŞ TAHMİN TABLOSU
# ============================================================
cm = confusion_matrix(
    y_test,
    y_pred,
    labels=np.arange(N_CLASSES)
)

class_rows = []

for i, cls in enumerate(CLASS_NAMES):
    test_count = int(cm[i, :].sum())
    correct_count = int(cm[i, i])
    wrong_count = int(test_count - correct_count)

    class_rows.append({
        "Gerçek sınıf": cls,
        "Test toplam": test_count,
        "Doğru tahmin": correct_count,
        "Yanlış tahmin": wrong_count,
        "Sınıf doğruluğu (%)": round(100 * correct_count / max(test_count, 1), 2)
    })

class_summary_df = pd.DataFrame(class_rows)

total_row = pd.DataFrame([{
    "Gerçek sınıf": "Toplam",
    "Test toplam": int(cm.sum()),
    "Doğru tahmin": int(np.trace(cm)),
    "Yanlış tahmin": int(cm.sum() - np.trace(cm)),
    "Sınıf doğruluğu (%)": round(100 * np.trace(cm) / max(cm.sum(), 1), 2)
}])

class_summary_df = pd.concat(
    [class_summary_df, total_row],
    ignore_index=True
)

print("\n===================================================")
print("SINIF BAZLI DOĞRU / YANLIŞ TAHMİN TABLOSU")
print("===================================================")
display(class_summary_df)

class_summary_df.to_csv(CLASS_SUMMARY_PATH, index=False)
print("\nSınıf bazlı sonuç tablosu kaydedildi:")
print(CLASS_SUMMARY_PATH)

# ============================================================
# 10) SINIFLAR ARASI KARIŞIKLIK TABLOSU
# ============================================================
confusion_detail_df = pd.DataFrame(
    cm,
    index=[f"Gerçek: {cls}" for cls in CLASS_NAMES],
    columns=[f"Tahmin: {cls}" for cls in CLASS_NAMES]
)

print("\n===================================================")
print("SINIFLAR ARASI KARIŞIKLIK TABLOSU")
print("Satırlar gerçek sınıfı, sütunlar tahmin edilen sınıfı gösterir.")
print("===================================================")
display(confusion_detail_df)

confusion_detail_df.to_csv(CONFUSION_DETAIL_PATH)
print("\nSınıflar arası karışıklık tablosu kaydedildi:")
print(CONFUSION_DETAIL_PATH)

# ============================================================
# 11) HER TEST GÖRÜNTÜSÜ İÇİN DETAYLI TAHMİN TABLOSU
# ============================================================
test_paths = [item[0] for item in test_items]

detail_rows = []

for idx in range(len(y_test)):
    true_label = int(y_test[idx])
    pred_label = int(y_pred[idx])

    true_class = IDX_TO_CLASS[true_label]
    pred_class = IDX_TO_CLASS[pred_label]

    max_prob = float(np.max(y_proba[idx]))
    true_prob = float(y_proba[idx, true_label])
    pred_prob = float(y_proba[idx, pred_label])

    detail_rows.append({
        "Dosya yolu": test_paths[idx],
        "Dosya adı": os.path.basename(test_paths[idx]),
        "Gerçek sınıf": true_class,
        "Tahmin edilen sınıf": pred_class,
        "Sonuç": "Doğru" if true_label == pred_label else "Yanlış",
        "Tahmin güveni": round(max_prob, 4),
        "Gerçek sınıf olasılığı": round(true_prob, 4),
        "Tahmin edilen sınıf olasılığı": round(pred_prob, 4)
    })

detail_df = pd.DataFrame(detail_rows)

print("\n===================================================")
print("GÖRÜNTÜ BAZLI DETAYLI TEST SONUÇLARI")
print("===================================================")
display(detail_df)

detail_df.to_csv(DETAIL_TEST_RESULTS_PATH, index=False)
print("\nGörüntü bazlı detaylı test sonuçları kaydedildi:")
print(DETAIL_TEST_RESULTS_PATH)

wrong_detail_df = detail_df[detail_df["Sonuç"] == "Yanlış"].copy()

print("\n===================================================")
print("SADECE YANLIŞ TAHMİN EDİLEN TEST GÖRÜNTÜLERİ")
print("===================================================")
display(wrong_detail_df)

# ============================================================
# 12) CLASSIFICATION REPORT
# ============================================================
print("\n===================================================")
print("CLASSIFICATION REPORT")
print("===================================================")
print(
    classification_report(
        y_test,
        y_pred,
        target_names=CLASS_NAMES,
        digits=4,
        zero_division=0
    )
)

# ============================================================
# 13) CONFUSION MATRIX GRAFİĞİ
# ============================================================
plt.figure(figsize=(8, 6))

disp = ConfusionMatrixDisplay(
    confusion_matrix=cm,
    display_labels=CLASS_NAMES
)

disp.plot(
    xticks_rotation=35,
    values_format="d",
    cmap="Blues"
)

plt.title("Klasik SVM - Confusion Matrix")
plt.grid(False)
plt.show()

# ============================================================
# 14) ROC EĞRİLERİ
# ============================================================
y_test_bin = label_binarize(
    y_test,
    classes=np.arange(N_CLASSES)
)

plt.figure(figsize=(8, 6))

for i, cls in enumerate(CLASS_NAMES):
    try:
        fpr, tpr, _ = roc_curve(
            y_test_bin[:, i],
            y_proba[:, i]
        )
        auc_i = auc(fpr, tpr)
        plt.plot(fpr, tpr, label=f"{cls} | AUC={auc_i:.4f}")
    except Exception:
        pass

plt.plot([0, 1], [0, 1], linestyle="--")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("Klasik SVM - ROC Eğrileri")
plt.legend()
plt.grid(True)
plt.show()

# ============================================================
# 15) MAKALE / SLAYT İÇİN KISA ÖZET
# ============================================================
print("\n===================================================")
print("SLAYT İÇİN KISA ÖZET")
print("===================================================")

print(f"""
Klasik SVM karşılaştırma modeli, ilk yüklenen daha iyi sonuç veren modeldeki
klasik SVM baseline akışıyla uyumlu olacak şekilde yeniden düzenlenmiştir.
MR görüntülerinden EfficientNet-B0 ile 1280 boyutlu derin öznitelikler çıkarılmış,
bu öznitelikler önce StandardScaler ile standartlaştırılmış, ardından PCA ile
{X_train_pca.shape[1]} boyuta indirgenmiştir. PCA sonrası öznitelikler, güçlü
modelde olduğu gibi Pipeline içinde yeniden StandardScaler işleminden geçirilmiş
ve RBF çekirdekli SVM sınıflandırıcısına verilmiştir.

SVM hiperparametreleri sabit verilmemiş; C ve gamma değerleri 5 katlı
StratifiedKFold çapraz doğrulama ve Macro-F1 ölçütüyle GridSearchCV kullanılarak
seçilmiştir. En iyi parametreler: {best_params}; CV Macro-F1: {best_cv_macro_f1:.4f}.

Bu modelde kuantum çekirdek, qubit, quantum feature map veya gated fusion
kullanılmamıştır. Nihai sınıflandırma tamamen klasik SVM yapısı ile
gerçekleştirilmiştir.

Test kümesindeki toplam {test_total} görüntünün {correct_total} tanesi doğru,
{wrong_total} tanesi yanlış sınıflandırılmıştır. Genel test doğruluğu
{accuracy:.4f}, Macro-F1 değeri {macro_f1:.4f}, Weighted-F1 değeri
{weighted_f1:.4f} ve Macro-AUC değeri {macro_auc:.4f} olarak elde edilmiştir.
""")
```
