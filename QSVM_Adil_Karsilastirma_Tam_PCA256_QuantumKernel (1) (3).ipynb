# Adil QSVM Modeli

Bu notebook klasik SVM çalıştırmadan, klasik SVM ile aynı ön işleme hattını kullanan adil bir QSVM modeli kurar.

Akış: EfficientNet-B0 → StandardScaler → PCA(256) → StandardScaler → Quantum Feature Map → Quantum Kernel → QSVM.


```python
# ============================================================
# ADİL QSVM MODELİ
# Klasik SVM yapısına adil karşılaştırma için düzenlenmiş QSVM
#
# Ana akış:
# MR görüntüsü
# -> EfficientNet-B0 özellik çıkarımı, 1280 özellik
# -> StandardScaler
# -> PCA(256)
# -> StandardScaler
# -> 256 PCA bileşeninin tamamını kullanan Quantum Feature Map
# -> Quantum Kernel
# -> SVC(kernel="precomputed")
#
# Not:
# Bu kod klasik SVM çalıştırmaz. Adil karşılaştırma için klasik SVM kodu da
# aynı split, aynı EfficientNet özellikleri, aynı scaler ve aynı PCA(256)
# ile çalıştırılmalıdır.
# ============================================================

# ============================================================
# 0) COLAB / KURULUM
# ============================================================
AUTO_INSTALL = True
MOUNT_DRIVE = True

import sys
import subprocess

if AUTO_INSTALL:
    packages = [
        "torchvision",
        "scikit-learn",
        "pandas",
        "tqdm",
        "matplotlib",
        "joblib",
        "openpyxl",
        "pennylane",
    ]
    subprocess.check_call([sys.executable, "-m", "pip", "install", "-q"] + packages)

try:
    if MOUNT_DRIVE:
        from google.colab import drive
        drive.mount("/content/drive")
except Exception:
    print("Google Drive bağlanmadı veya Colab dışında çalışıyorsunuz.")

# ============================================================
# 1) IMPORTLAR
# ============================================================
import os
import math
import random
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
from PIL import Image
from tqdm.auto import tqdm

import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader

import torchvision.transforms as T
import torchvision.models as models

import matplotlib.pyplot as plt
import joblib
import pennylane as qml

from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.preprocessing import StandardScaler, label_binarize
from sklearn.decomposition import PCA
from sklearn.svm import SVC
from sklearn.metrics import (
    accuracy_score,
    f1_score,
    roc_auc_score,
    classification_report,
    confusion_matrix,
    ConfusionMatrixDisplay,
    roc_curve,
    auc,
)

# ============================================================
# 2) GENEL AYARLAR
# ============================================================
SEED = 42

def seed_everything(seed=42):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    torch.backends.cudnn.deterministic = False
    torch.backends.cudnn.benchmark = True

seed_everything(SEED)

DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Kullanılan cihaz:", DEVICE)

# Veri klasörü. Gerekirse kendi Drive yoluna göre düzenle.
DATA_ROOT = "/content/drive/MyDrive/Beyin_MR_Kuantum_Makale/Data"

CLASS_NAMES = [
    "glioma_tumor",
    "meningioma_tumor",
    "no_tumor",
    "pituitary_tumor",
]

# Veri setindeki olası klasör isimleri otomatik tanınır.
CLASS_ALIASES = {
    "glioma_tumor": ["glioma_tumor", "glioma", "glioma tumor"],
    "meningioma_tumor": ["meningioma_tumor", "meningioma", "meningioma tumor"],
    "no_tumor": ["no_tumor", "notumor", "no tumor", "normal"],
    "pituitary_tumor": ["pituitary_tumor", "pituitary", "pituitary tumor"],
}

CLASS_TO_IDX = {cls: i for i, cls in enumerate(CLASS_NAMES)}
IDX_TO_CLASS = {i: cls for cls, i in CLASS_TO_IDX.items()}
N_CLASSES = len(CLASS_NAMES)

print("Sınıf eşlemesi:", CLASS_TO_IDX)

IMG_SIZE = 224
TEST_SIZE = 0.20
FEATURE_BATCH_SIZE = 64

# Adil karşılaştırma için klasik SVM ile aynı PCA boyutu.
# QSVM bu 256 bileşenin tamamını data re-uploading ile quantum devreye taşır.
PCA_COMPONENTS = 256

# QSVM quantum ayarları.
# 8 qubit -> state boyutu 256. PCA(256), 8 qubit üzerinde 32 re-uploading katmanı ile işlenir.
N_QUBITS = 8
QSVM_FEATURES_USED = PCA_COMPONENTS
N_REUPLOAD_LAYERS = math.ceil(QSVM_FEATURES_USED / N_QUBITS)
TOTAL_Q_FEATURES = N_REUPLOAD_LAYERS * N_QUBITS

# Quantum kernel hiperparametreleri.
# Daha fazla scale denemesi daha uzun sürer ama bazen daha iyi sonuç verir.
FEATURE_SCALE_GRID = [0.35, 0.50, 0.75, 1.00, 1.25]
C_GRID = [0.3, 1, 3, 10, 30, 100]
CV_SPLITS = 5

# Açılar için stabilizasyon. PCA sonrası aşırı değerleri quantum açılarda sınırlıyoruz.
ANGLE_CLIP = 3.0
ANGLE_MULTIPLIER = np.pi / ANGLE_CLIP

FORCE_NEW_SPLIT = True
FORCE_REEXTRACT_IMAGE_FEATURES = False
FORCE_RECOMPUTE_PREPROCESS = False
FORCE_RECOMPUTE_QUANTUM_STATES = False

CACHE_DIR = "/content/drive/MyDrive/Beyin_MR_Kuantum_Makale/cache_fair_qsvm_full_pca256"
os.makedirs(CACHE_DIR, exist_ok=True)

TRAIN_SPLIT_PATH = os.path.join(CACHE_DIR, "train_split_80_20.csv")
TEST_SPLIT_PATH = os.path.join(CACHE_DIR, "test_split_80_20.csv")
FEATURE_CACHE_PATH = os.path.join(CACHE_DIR, "efficientnet_b0_features_80_20.joblib")
PREPROCESS_CACHE_PATH = os.path.join(CACHE_DIR, "scaler_pca256_postscaler.joblib")
MODEL_SAVE_PATH = os.path.join(CACHE_DIR, "fair_qsvm_model.joblib")
RESULTS_EXCEL_PATH = os.path.join(CACHE_DIR, "fair_qsvm_test_predictions.xlsx")
CONFUSION_MATRIX_PATH = os.path.join(CACHE_DIR, "fair_qsvm_confusion_matrix.png")
ROC_CURVE_PATH = os.path.join(CACHE_DIR, "fair_qsvm_roc_curves.png")

print("\nQSVM adil ayar özeti")
print("PCA_COMPONENTS:", PCA_COMPONENTS)
print("QSVM_FEATURES_USED:", QSVM_FEATURES_USED)
print("N_QUBITS:", N_QUBITS)
print("N_REUPLOAD_LAYERS:", N_REUPLOAD_LAYERS)
print("CV_SPLITS:", CV_SPLITS)
print("FEATURE_SCALE_GRID:", FEATURE_SCALE_GRID)
print("C_GRID:", C_GRID)

# ============================================================
# 3) VERİ KLASÖRLERİNİ BULMA VE %80/%20 CLASSWISE SPLIT
# ============================================================
IMG_EXTENSIONS = (".jpg", ".jpeg", ".png", ".bmp", ".webp")

def normalize_name(name):
    return name.lower().strip().replace("-", "_").replace(" ", "_")

def find_class_files(data_root, class_names, class_aliases):
    if not os.path.exists(data_root):
        raise FileNotFoundError(f"DATA_ROOT bulunamadı: {data_root}")

    alias_to_class = {}
    for cls in class_names:
        for a in class_aliases.get(cls, [cls]):
            alias_to_class[normalize_name(a)] = cls

    class_files = {cls: [] for cls in class_names}
    matched_dirs = {cls: [] for cls in class_names}

    for root, dirs, files in os.walk(data_root):
        base = normalize_name(os.path.basename(root))
        if base in alias_to_class:
            cls = alias_to_class[base]
            img_files = [
                os.path.join(root, f)
                for f in files
                if f.lower().endswith(IMG_EXTENSIONS)
            ]
            if img_files:
                class_files[cls].extend(img_files)
                matched_dirs[cls].append(root)

    # Tekrarlı dosyaları temizle.
    for cls in class_names:
        class_files[cls] = sorted(list(dict.fromkeys(class_files[cls])))

    print("\nBulunan sınıf klasörleri ve görüntü sayıları:")
    for cls in class_names:
        print(f"{cls}: {len(class_files[cls])} görüntü")
        for d in matched_dirs[cls]:
            print("  -", d)
        if len(class_files[cls]) == 0:
            aliases = class_aliases.get(cls, [cls])
            raise ValueError(
                f"Bu sınıf için görüntü bulunamadı: {cls}. "
                f"Aranan olası klasör adları: {aliases}"
            )

    return class_files

def build_classwise_split(data_root, class_names, class_aliases, test_size=0.20, seed=42):
    class_files = find_class_files(data_root, class_names, class_aliases)
    train_items, test_items, all_items = [], [], []

    for cls in class_names:
        y = CLASS_TO_IDX[cls]
        files = class_files[cls]

        tr_files, te_files = train_test_split(
            files,
            test_size=test_size,
            random_state=seed,
            shuffle=True,
        )

        train_items.extend([(p, y, cls) for p in tr_files])
        test_items.extend([(p, y, cls) for p in te_files])
        all_items.extend([(p, y, cls) for p in files])

    rng = np.random.default_rng(seed)
    rng.shuffle(train_items)
    rng.shuffle(test_items)

    return train_items, test_items, all_items

def save_split(items, path):
    df = pd.DataFrame(items, columns=["path", "label", "class_name"])
    df.to_csv(path, index=False)

def load_split(path):
    df = pd.read_csv(path)
    return list(df[["path", "label", "class_name"]].itertuples(index=False, name=None))

def show_split_table(train_items, test_items):
    rows = []
    for cls in CLASS_NAMES:
        tr = sum(1 for _, _, c in train_items if c == cls)
        te = sum(1 for _, _, c in test_items if c == cls)
        total = tr + te
        rows.append({
            "Sınıf": cls,
            "Eğitim": tr,
            "Test": te,
            "Toplam": total,
            "Test Oranı": round(te / total, 4) if total else np.nan,
        })
    df = pd.DataFrame(rows)
    print("\nSınıf bazlı eğitim/test dağılımı:")
    print(df)
    print("\nToplam eğitim:", len(train_items))
    print("Toplam test:", len(test_items))
    return df

if (not FORCE_NEW_SPLIT) and os.path.exists(TRAIN_SPLIT_PATH) and os.path.exists(TEST_SPLIT_PATH):
    print("\nMevcut split dosyaları yükleniyor.")
    train_items = load_split(TRAIN_SPLIT_PATH)
    test_items = load_split(TEST_SPLIT_PATH)
else:
    print("\nHer sınıftan %80 eğitim / %20 test olacak şekilde yeni split oluşturuluyor.")
    train_items, test_items, all_items = build_classwise_split(
        DATA_ROOT,
        CLASS_NAMES,
        CLASS_ALIASES,
        test_size=TEST_SIZE,
        seed=SEED,
    )
    save_split(train_items, TRAIN_SPLIT_PATH)
    save_split(test_items, TEST_SPLIT_PATH)

split_df = show_split_table(train_items, test_items)

# ============================================================
# 4) EFFICIENTNET-B0 ÖZELLİK ÇIKARIMI
# ============================================================
feature_transform = T.Compose([
    T.Resize((IMG_SIZE, IMG_SIZE)),
    T.ToTensor(),
    T.Normalize(
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225],
    ),
])

class ImagePathDataset(Dataset):
    def __init__(self, items, transform):
        self.items = items
        self.transform = transform

    def __len__(self):
        return len(self.items)

    def __getitem__(self, idx):
        path, y, _ = self.items[idx]
        img = Image.open(path).convert("RGB")
        x = self.transform(img)
        return x, int(y), path

def build_efficientnet_b0_feature_extractor():
    weights = models.EfficientNet_B0_Weights.DEFAULT
    model = models.efficientnet_b0(weights=weights)
    model.classifier = nn.Identity()
    model = model.to(DEVICE)
    model.eval()
    for p in model.parameters():
        p.requires_grad = False
    return model, 1280

feature_extractor, RAW_FEATURE_DIM = build_efficientnet_b0_feature_extractor()
print("\nFeature extractor: EfficientNet-B0")
print("Ham özellik boyutu:", RAW_FEATURE_DIM)

@torch.no_grad()
def extract_image_features(items, split_name):
    dataset = ImagePathDataset(items, feature_transform)
    loader = DataLoader(
        dataset,
        batch_size=FEATURE_BATCH_SIZE,
        shuffle=False,
        num_workers=2,
        pin_memory=(DEVICE.type == "cuda"),
    )

    features, labels, paths = [], [], []
    for x, y, p in tqdm(loader, desc=f"{split_name} EfficientNet-B0 özellikleri çıkarılıyor"):
        x = x.to(DEVICE)
        out = feature_extractor(x)
        out = out.detach().cpu().numpy().astype(np.float32)
        features.append(out)
        labels.extend(y.numpy().astype(int).tolist())
        paths.extend(list(p))

    X = np.concatenate(features, axis=0)
    y = np.array(labels, dtype=np.int64)
    return X, y, paths

def cache_paths_match(cached_paths, current_items):
    current_paths = [p for p, _, _ in current_items]
    return list(cached_paths) == list(current_paths)

if (not FORCE_REEXTRACT_IMAGE_FEATURES) and os.path.exists(FEATURE_CACHE_PATH):
    cache = joblib.load(FEATURE_CACHE_PATH)
    if cache_paths_match(cache["train_paths"], train_items) and cache_paths_match(cache["test_paths"], test_items):
        print("\nEfficientNet özellikleri cache üzerinden yükleniyor.")
        X_train_raw = cache["X_train_raw"]
        y_train = cache["y_train"]
        train_paths = cache["train_paths"]
        X_test_raw = cache["X_test_raw"]
        y_test = cache["y_test"]
        test_paths = cache["test_paths"]
    else:
        print("\nCache split ile uyuşmuyor. Özellikler yeniden çıkarılıyor.")
        X_train_raw, y_train, train_paths = extract_image_features(train_items, "Eğitim")
        X_test_raw, y_test, test_paths = extract_image_features(test_items, "Test")
        joblib.dump({
            "X_train_raw": X_train_raw,
            "y_train": y_train,
            "train_paths": train_paths,
            "X_test_raw": X_test_raw,
            "y_test": y_test,
            "test_paths": test_paths,
        }, FEATURE_CACHE_PATH)
else:
    print("\nEfficientNet özellikleri çıkarılıyor.")
    X_train_raw, y_train, train_paths = extract_image_features(train_items, "Eğitim")
    X_test_raw, y_test, test_paths = extract_image_features(test_items, "Test")
    joblib.dump({
        "X_train_raw": X_train_raw,
        "y_train": y_train,
        "train_paths": train_paths,
        "X_test_raw": X_test_raw,
        "y_test": y_test,
        "test_paths": test_paths,
    }, FEATURE_CACHE_PATH)

print("\nHam özellik şekilleri:")
print("X_train_raw:", X_train_raw.shape)
print("X_test_raw :", X_test_raw.shape)

# ============================================================
# 5) SCALER + PCA(256) + SCALER
# ============================================================
if (not FORCE_RECOMPUTE_PREPROCESS) and os.path.exists(PREPROCESS_CACHE_PATH):
    pp = joblib.load(PREPROCESS_CACHE_PATH)
    if pp.get("train_shape") == X_train_raw.shape and pp.get("pca_components_requested") == PCA_COMPONENTS:
        print("\nScaler + PCA cache üzerinden yükleniyor.")
        raw_scaler = pp["raw_scaler"]
        pca = pp["pca"]
        post_scaler = pp["post_scaler"]
        X_train_pca = pp["X_train_pca"]
        X_test_pca = pp["X_test_pca"]
    else:
        print("\nPreprocess cache uyuşmuyor. Yeniden hesaplanıyor.")
        raw_scaler = None
else:
    raw_scaler = None

if raw_scaler is None:
    n_components = min(PCA_COMPONENTS, X_train_raw.shape[0] - 1, X_train_raw.shape[1])
    if n_components != PCA_COMPONENTS:
        print(f"Uyarı: PCA_COMPONENTS {PCA_COMPONENTS} yerine {n_components} olarak ayarlandı.")

    raw_scaler = StandardScaler()
    X_train_scaled = raw_scaler.fit_transform(X_train_raw)
    X_test_scaled = raw_scaler.transform(X_test_raw)

    pca = PCA(n_components=n_components, random_state=SEED, svd_solver="randomized")
    X_train_pca_raw = pca.fit_transform(X_train_scaled)
    X_test_pca_raw = pca.transform(X_test_scaled)

    post_scaler = StandardScaler()
    X_train_pca = post_scaler.fit_transform(X_train_pca_raw).astype(np.float32)
    X_test_pca = post_scaler.transform(X_test_pca_raw).astype(np.float32)

    joblib.dump({
        "raw_scaler": raw_scaler,
        "pca": pca,
        "post_scaler": post_scaler,
        "X_train_pca": X_train_pca,
        "X_test_pca": X_test_pca,
        "train_shape": X_train_raw.shape,
        "pca_components_requested": PCA_COMPONENTS,
        "pca_components_actual": n_components,
    }, PREPROCESS_CACHE_PATH)

explained_var = float(np.sum(pca.explained_variance_ratio_))
print("\nPCA sonrası şekiller:")
print("X_train_pca:", X_train_pca.shape)
print("X_test_pca :", X_test_pca.shape)
print(f"PCA açıklanan varyans toplamı: {explained_var:.4f}")

# Quantum açı girdilerini stabilize et.
def to_quantum_angles(X):
    Xq = X[:, :QSVM_FEATURES_USED]
    if Xq.shape[1] < QSVM_FEATURES_USED:
        pad = np.zeros((Xq.shape[0], QSVM_FEATURES_USED - Xq.shape[1]), dtype=np.float32)
        Xq = np.concatenate([Xq, pad], axis=1)
    Xq = np.clip(Xq, -ANGLE_CLIP, ANGLE_CLIP) * ANGLE_MULTIPLIER
    return Xq.astype(np.float32)

X_train_q = to_quantum_angles(X_train_pca)
X_test_q = to_quantum_angles(X_test_pca)

print("\nQuantum giriş şekilleri:")
print("X_train_q:", X_train_q.shape)
print("X_test_q :", X_test_q.shape)

# ============================================================
# 6) QUANTUM FEATURE MAP VE KERNEL
# ============================================================
try:
    qdev = qml.device("lightning.qubit", wires=N_QUBITS)
    QDEVICE_NAME = "lightning.qubit"
except Exception:
    qdev = qml.device("default.qubit", wires=N_QUBITS)
    QDEVICE_NAME = "default.qubit"

print("\nQuantum device:", QDEVICE_NAME)
print("Quantum state boyutu:", 2 ** N_QUBITS)

@qml.qnode(qdev, interface="numpy", diff_method=None)
def quantum_feature_state(x, feature_scale=1.0):
    # Başlangıç süperpozisyonu
    for w in range(N_QUBITS):
        qml.Hadamard(wires=w)

    # PCA(256) -> 8 qubit üzerinde 32 kez data re-uploading
    for layer in range(N_REUPLOAD_LAYERS):
        offset = layer * N_QUBITS
        for w in range(N_QUBITS):
            angle = feature_scale * x[offset + w]
            qml.RY(angle, wires=w)
            qml.RZ(0.5 * angle, wires=w)

        # Ring entanglement
        for w in range(N_QUBITS - 1):
            qml.CNOT(wires=[w, w + 1])
        qml.CNOT(wires=[N_QUBITS - 1, 0])

    return qml.state()

def pad_quantum_input(x):
    out = np.zeros(TOTAL_Q_FEATURES, dtype=np.float64)
    n = min(len(x), QSVM_FEATURES_USED)
    out[:n] = x[:n]
    return out

def quantum_state_cache_path(split_name, scale):
    safe_scale = str(scale).replace(".", "p")
    return os.path.join(
        CACHE_DIR,
        f"qstates_{split_name}_nq{N_QUBITS}_features{QSVM_FEATURES_USED}_layers{N_REUPLOAD_LAYERS}_scale{safe_scale}.npy",
    )

def compute_quantum_states(X, split_name, feature_scale):
    cache_path = quantum_state_cache_path(split_name, feature_scale)
    if (not FORCE_RECOMPUTE_QUANTUM_STATES) and os.path.exists(cache_path):
        states = np.load(cache_path)
        if states.shape[0] == X.shape[0] and states.shape[1] == 2 ** N_QUBITS:
            print(f"{split_name} quantum state cache yüklendi: {cache_path}")
            return states

    states = np.empty((X.shape[0], 2 ** N_QUBITS), dtype=np.complex128)
    for i in tqdm(range(X.shape[0]), desc=f"{split_name} quantum state çıkarımı | scale={feature_scale}"):
        x_pad = pad_quantum_input(X[i])
        state = quantum_feature_state(x_pad, feature_scale=feature_scale)
        state = np.asarray(state, dtype=np.complex128)
        norm = np.linalg.norm(state)
        if norm > 0:
            state = state / norm
        states[i] = state

    np.save(cache_path, states)
    return states

def quantum_kernel_from_states(A, B=None):
    if B is None:
        K = np.abs(A @ A.conj().T) ** 2
        K = np.real(K)
        K = 0.5 * (K + K.T)
        np.fill_diagonal(K, 1.0)
        K = np.clip(K, 0.0, 1.0)
        K += 1e-8 * np.eye(K.shape[0])
        return K.astype(np.float64)
    K = np.abs(A @ B.conj().T) ** 2
    K = np.real(K)
    K = np.clip(K, 0.0, 1.0)
    return K.astype(np.float64)

# ============================================================
# 7) SADECE QUANTUM KERNEL İLE QSVM HİPERPARAMETRE ARAMASI
# ============================================================
print("\n===================================================")
print("ADİL QSVM EĞİTİMİ BAŞLIYOR")
print("===================================================")
print("Klasik SVM/kernel kullanılmıyor.")
print("Tüm PCA(256) bileşenleri quantum feature map'e data re-uploading ile aktarılıyor.")

cv = StratifiedKFold(n_splits=CV_SPLITS, shuffle=True, random_state=SEED)
search_rows = []
best_score = -1.0
best_config = None
best_train_states = None
best_test_states = None
best_K_train = None
best_K_test = None

for feature_scale in FEATURE_SCALE_GRID:
    print("\n---------------------------------------------------")
    print(f"Quantum kernel hesaplanıyor | feature_scale={feature_scale}")
    print("---------------------------------------------------")

    train_states = compute_quantum_states(X_train_q, "train", feature_scale)
    test_states = compute_quantum_states(X_test_q, "test", feature_scale)

    K_train = quantum_kernel_from_states(train_states)

    for C in C_GRID:
        fold_scores = []
        fold_accs = []

        for fold, (tr_idx, val_idx) in enumerate(cv.split(K_train, y_train), start=1):
            K_tr = K_train[np.ix_(tr_idx, tr_idx)]
            K_val = K_train[np.ix_(val_idx, tr_idx)]
            y_tr = y_train[tr_idx]
            y_val = y_train[val_idx]

            model = SVC(
                kernel="precomputed",
                C=C,
                class_weight="balanced",
                decision_function_shape="ovr",
                probability=False,
                random_state=SEED,
            )
            model.fit(K_tr, y_tr)
            pred = model.predict(K_val)
            fold_f1 = f1_score(y_val, pred, average="macro", zero_division=0)
            fold_acc = accuracy_score(y_val, pred)
            fold_scores.append(fold_f1)
            fold_accs.append(fold_acc)

        mean_f1 = float(np.mean(fold_scores))
        std_f1 = float(np.std(fold_scores))
        mean_acc = float(np.mean(fold_accs))

        row = {
            "feature_scale": feature_scale,
            "C": C,
            "cv_macro_f1_mean": mean_f1,
            "cv_macro_f1_std": std_f1,
            "cv_accuracy_mean": mean_acc,
            "n_qubits": N_QUBITS,
            "features_used": QSVM_FEATURES_USED,
            "reupload_layers": N_REUPLOAD_LAYERS,
        }
        search_rows.append(row)
        print(f"scale={feature_scale} | C={C} | CV Macro-F1={mean_f1:.4f} ± {std_f1:.4f} | Acc={mean_acc:.4f}")

        if mean_f1 > best_score:
            best_score = mean_f1
            best_config = row.copy()
            best_train_states = train_states
            best_test_states = test_states
            best_K_train = K_train

print("\nEn iyi QSVM ayarı:")
print(best_config)

# Final test kernel sadece en iyi scale için hesaplanır.
best_K_test = quantum_kernel_from_states(best_test_states, best_train_states)

# Final model tüm eğitim setiyle eğitilir.
qsvm_model = SVC(
    kernel="precomputed",
    C=best_config["C"],
    class_weight="balanced",
    decision_function_shape="ovr",
    probability=False,
    random_state=SEED,
)
qsvm_model.fit(best_K_train, y_train)

joblib.dump({
    "model": qsvm_model,
    "raw_scaler": raw_scaler,
    "pca": pca,
    "post_scaler": post_scaler,
    "class_names": CLASS_NAMES,
    "class_to_idx": CLASS_TO_IDX,
    "idx_to_class": IDX_TO_CLASS,
    "pca_components": PCA_COMPONENTS,
    "qsvm_features_used": QSVM_FEATURES_USED,
    "n_qubits": N_QUBITS,
    "n_reupload_layers": N_REUPLOAD_LAYERS,
    "feature_scale_grid": FEATURE_SCALE_GRID,
    "C_grid": C_GRID,
    "best_config": best_config,
    "search_results": pd.DataFrame(search_rows),
    "explained_variance": explained_var,
    "note": "Sadece quantum kernel tabanlı QSVM. Klasik SVM/kernel kullanılmadı.",
}, MODEL_SAVE_PATH)

print("\nModel kaydedildi:", MODEL_SAVE_PATH)

# ============================================================
# 8) TEST DEĞERLENDİRMESİ
# ============================================================
def safe_macro_auc_from_decision(y_true, decision_scores):
    try:
        if decision_scores.ndim == 1:
            return np.nan
        return roc_auc_score(
            y_true,
            decision_scores,
            multi_class="ovr",
            average="macro",
        )
    except Exception:
        return np.nan

y_pred = qsvm_model.predict(best_K_test)
try:
    y_decision = qsvm_model.decision_function(best_K_test)
except Exception:
    y_decision = None

accuracy = accuracy_score(y_test, y_pred)
macro_f1 = f1_score(y_test, y_pred, average="macro", zero_division=0)
weighted_f1 = f1_score(y_test, y_pred, average="weighted", zero_division=0)
macro_auc = safe_macro_auc_from_decision(y_test, y_decision) if y_decision is not None else np.nan

print("\n===================================================")
print("ADİL QSVM TEST SONUÇLARI")
print("===================================================")
print(f"Accuracy     : {accuracy:.4f}")
print(f"Macro-F1     : {macro_f1:.4f}")
print(f"Weighted-F1  : {weighted_f1:.4f}")
print(f"Macro-AUC    : {macro_auc:.4f}" if not np.isnan(macro_auc) else "Macro-AUC    : hesaplanamadı")
print("En iyi ayar  :", best_config)

print("\nClassification report:")
print(classification_report(y_test, y_pred, target_names=CLASS_NAMES, zero_division=0))

# Sınıf bazlı doğru/yanlış tablosu
rows = []
for i, cls in enumerate(CLASS_NAMES):
    mask = (y_test == i)
    total = int(mask.sum())
    correct = int((y_pred[mask] == i).sum())
    wrong = total - correct
    acc_cls = correct / total if total > 0 else np.nan
    rows.append({
        "Sınıf": cls,
        "Test Sayısı": total,
        "Doğru": correct,
        "Yanlış": wrong,
        "Sınıf Doğruluğu": acc_cls,
    })

class_summary_df = pd.DataFrame(rows)
print("\nSınıf bazlı doğru / yanlış tahmin tablosu:")
print(class_summary_df)

# Detaylı tahmin tablosu
pred_df = pd.DataFrame({
    "path": test_paths,
    "true_label": y_test,
    "true_class": [IDX_TO_CLASS[int(y)] for y in y_test],
    "pred_label": y_pred,
    "pred_class": [IDX_TO_CLASS[int(y)] for y in y_pred],
    "correct": y_test == y_pred,
})

search_df = pd.DataFrame(search_rows).sort_values("cv_macro_f1_mean", ascending=False)

with pd.ExcelWriter(RESULTS_EXCEL_PATH) as writer:
    split_df.to_excel(writer, sheet_name="split_distribution", index=False)
    search_df.to_excel(writer, sheet_name="qsvm_cv_search", index=False)
    class_summary_df.to_excel(writer, sheet_name="class_summary", index=False)
    pred_df.to_excel(writer, sheet_name="test_predictions", index=False)

print("\nSonuç Excel dosyası kaydedildi:", RESULTS_EXCEL_PATH)

# Confusion matrix
cm = confusion_matrix(y_test, y_pred, labels=list(range(N_CLASSES)))
fig, ax = plt.subplots(figsize=(7, 6))
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=CLASS_NAMES)
disp.plot(ax=ax, xticks_rotation=45, cmap="Blues", colorbar=False)
plt.title("Fair QSVM Confusion Matrix")
plt.tight_layout()
plt.savefig(CONFUSION_MATRIX_PATH, dpi=200)
plt.show()
print("Confusion matrix kaydedildi:", CONFUSION_MATRIX_PATH)

# ROC curves, decision_function ile
if y_decision is not None and getattr(y_decision, "ndim", 1) == 2:
    y_test_bin = label_binarize(y_test, classes=list(range(N_CLASSES)))
    plt.figure(figsize=(7, 6))
    for i, cls in enumerate(CLASS_NAMES):
        try:
            fpr, tpr, _ = roc_curve(y_test_bin[:, i], y_decision[:, i])
            roc_auc = auc(fpr, tpr)
            plt.plot(fpr, tpr, label=f"{cls} AUC={roc_auc:.3f}")
        except Exception:
            pass
    plt.plot([0, 1], [0, 1], "--", label="Random")
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title("Fair QSVM ROC Curves")
    plt.legend(fontsize=8)
    plt.tight_layout()
    plt.savefig(ROC_CURVE_PATH, dpi=200)
    plt.show()
    print("ROC eğrisi kaydedildi:", ROC_CURVE_PATH)
else:
    print("ROC eğrisi çizilemedi: decision_function uygun formatta değil.")

print("\nBitti.")
print("Bu QSVM kodu klasik kernel kullanmaz; adil karşılaştırma için klasik SVM de aynı split ve PCA(256) ile çalıştırılmalıdır.")
```
